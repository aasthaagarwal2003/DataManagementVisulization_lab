```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

# Ensure it displays
pio.renderers.default = "browser"   # use "notebook" if working

# Load dataset
df = pd.read_csv("Companies_dataset.csv")

# Drop unwanted column
if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)

# Clean ratings
df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')

# Remove null ratings
df_clean = df.dropna(subset=['ratings'])

# Define refined funnel stages
stages = [
    "Total Companies",
    "Valid Ratings",
    "Rating ≥ 3",
    "Rating ≥ 4",
    "Rating ≥ 4.5"
]

values = [
    len(df),
    len(df_clean),
    len(df_clean[df_clean['ratings'] >= 3]),
    len(df_clean[df_clean['ratings'] >= 4]),
    len(df_clean[df_clean['ratings'] >= 4.5])
]

# Create funnel chart
fig = px.funnel(
    x=values,
    y=stages,
    title="Refined Funnel Chart of Companies Based on Ratings"
)

# Improve layout
fig.update_layout(
    title_x=0.5,
    font=dict(size=14)
)

fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Companies_dataset.csv")

# Drop unwanted column
if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)

# Clean years column
df['years'] = df['years'].str.extract('(\d+)').astype(float)

# Drop missing values
df = df.dropna(subset=['years'])

# 🔥 Create year groups (VERY IMPORTANT)
bins = [0, 5, 10, 20, 50]
labels = ['0-5', '5-10', '10-20', '20+']

df['year_group'] = pd.cut(df['years'], bins=bins, labels=labels)

# Count companies in each group
year_counts = df['year_group'].value_counts().sort_index()

# Plot bar chart
plt.figure()
plt.bar(year_counts.index, year_counts.values)

plt.xlabel("Company Age Group (Years)")
plt.ylabel("Number of Companies")
plt.title("Company Distribution by Age")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Companies_dataset.csv")

# Drop unwanted column
if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)

# Clean ratings
df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')

# Clean years
df['years'] = df['years'].str.extract('(\d+)').astype(float)

# Drop missing values
df = df.dropna(subset=['years', 'ratings'])

# 🔥 Create bins (VERY IMPORTANT)
bins = [0, 5, 10, 20, 50]
labels = ['0-5', '5-10', '10-20', '20+']

df['year_group'] = pd.cut(df['years'], bins=bins, labels=labels)

# Group by bins
avg_rating = df.groupby('year_group')['ratings'].mean()

# Plot clean line chart
plt.figure()
plt.plot(avg_rating.index, avg_rating.values, marker='o')

plt.xlabel("Company Age Group (Years)")
plt.ylabel("Average Rating")
plt.title("Average Rating by Company Age Group")

plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Companies_dataset.csv")

# Drop unwanted column
if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)

# 🔥 Clean employees column
df['employees'] = df['employees'].str.lower().str.strip()

# Simplify categories (group similar ones)
def categorize(emp):
    if '0-50' in emp or '1-50' in emp:
        return '0-50'
    elif '50-200' in emp:
        return '50-200'
    elif '200-500' in emp:
        return '200-500'
    elif '500-1k' in emp or '500-1000' in emp:
        return '500-1k'
    elif '1k-5k' in emp:
        return '1k-5k'
    elif '5k-10k' in emp:
        return '5k-10k'
    else:
        return '10k+'

df['emp_group'] = df['employees'].apply(lambda x: categorize(str(x)))

# Count values
emp_counts = df['emp_group'].value_counts()

# Plot pie chart
plt.figure()
plt.pie(emp_counts, labels=emp_counts.index, autopct='%1.1f%%')

plt.title("Distribution of Companies by Employee Size")

plt.show()
```
